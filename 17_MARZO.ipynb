````python
Comida=["combo 1", "combo 2", "combo 3", "combo 4","papas", "jugos naturales","cerveza", "agua" "gaseosa"]
precios= [20000, 25000, 18000, 24500, 3000, 7000, 3500, 4000, 2500]
````

````python
for menu in range(len(Comida)):
  print(Comida[menu], "-", precios[menu])
````

````python
a=6
r="s"
user="s"
while user=="s":
  while r=="s":
    cantidad=input ("ingresa la cantidad del producto:")
    x=input( "Qué desea?")
    if x in Comida:
      cantidad=int(input(f"cuantas unidades de {x} quieres pedir:"))
      a=a+precios[Comida.index(x)]*cantidad
    else:
      x=input("Desea algo mas?")
      if x=="s": continue
      else: break
    user=input ("Hay más usuarios?")
print(f"Usted ha ganado ${a} en el día")
````

#FUNCION
Es un pedazo de codigo wue se ejecuta aparte. A veces se conoce como copiado dinamico.

Ventajas de implementar las funciones
1. Vuleve el codigo mas limpio
2. Elimina el codigo espaguetti
3. Establece un orden de busqueda
4. Permite reutilizacion del codigo

SINTAXIS


```
def nombre_de_la_funcion n( parametros)
  Aqui va todo lo que desees ejecutar
  return algo
```
Existen 4 tipos de funciones
1. Sin parametros, sin retorno
2. Sin parametro, con retorno
3. Con parametros, sin retorno
4. Con parametros, con retornos  


````python
#Sin parametro, sin retorno
def menu():
  Comida=["combo 1", "combo 2", "combo 3", "combo 4","papas", "jugos naturales","cerveza", "agua" "gaseosa"]
  precios= [20000, 25000, 18000, 24500, 3000, 7000, 3500, 4000, 2500]
  for menu in range(len(Comida)):
    print(Comida[menu], "-", precios[menu])
````

````python
#Sin parametro, sin retorno
def menu():
  # preci y comi son variables locales - precio y comida variables generales
  comi=["combo 1", "combo 2", "combo 3", "combo 4","papas", "jugos naturales","cerveza", "agua" "gaseosa"]
  preci= [20000, 25000, 18000, 24500, 3000, 7000, 3500, 4000, 2500]
  for menu in range(len(comi)):
    print(Comida[menu], "-", precios[menu])
````

````python
#Sin parametro, sin retorno
menu()
````
