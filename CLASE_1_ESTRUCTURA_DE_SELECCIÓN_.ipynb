1. IF/ELSE
Sólo tiene dos posibles respuestas (V) o (F)
Sintaxis
````
If (condición):
 aquí se ejecuta código **si** la condición es verdadera.

else:
 aquí de ejecuta el código **si** la condición es falsa.

````

`````python
lluvia=input("Está lloviendo? ")
if lluvia=="no":
  print("Te recomiendo ir al parque")
else:
  if lluvia=="un poquito":
    print("Te sugiro que si vas al parque lleva sombrilla ")
  else:
    if lluvia=="si":
      print("Cierra las ventanas")
    else:
      print("resguárdese")
`````

`````python
resp="Te recomiendo ir al parque" if lluvia=="no" else "Te sugiero que si vas al parque lleva sombrilla"  if lluvia=="un poquito" else "Cierra las ventanas" if lluvia=="si" else "resguárdese"
print(resp)
`````

String o caracteres - input

`````python
temp=float(input("Temperatura registrada"))
if temp>=27: print("Póngase algo fresco👗")
elif temp>=20 and temp<27: print("Abrígate 🧥")
elif temp>=16 and temp<20: print("Abrígate más 🧥🧣")
else: print("Está helado 🥶")
`````

##SI ternario (Easy hand if)
Se emplea cuando la estructura solo posee un if-else (recomendado).
Sintaxis
````
variable = (acción si la condición es verdadera) if (condeción) else (acción si la condición es falsa)
````
Sintaxis en java

````
variable = (codición)? (acción si la condición es verdadera) : (acción si la condición es falsa);
````

`````python
#Programa que me dija si eres mayor de edad
edad=int(input("Cuál es tu edad? "))
if edad>=18: print("Eres mayor de edad 😊")
else: print("Eres menor de edad 😞")
`````

`````python
resp="Eres mayor de edad 😊" if edad>=18 else "Eres menor de edad 😞"
print(resp)
`````
