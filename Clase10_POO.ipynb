`````python
# ── DESCRIPCIÓN DE LUGARES ────────────────────────────────

def pedir(msg, num=False):
    while True:
        v = input(f"  {msg}: ").strip()
        if not v:
            print("  ⚠ Campo obligatorio.")
            continue
        if num:
            try: return int(v)
            except: print("  ⚠ Ingresa un número."); continue
        return v

def pedir_lista(msg):
    print(f"\n  {msg} (deja en blanco para terminar)")
    items, i = [], 1
    while True:
        v = input(f"    [{i}] ").strip()
        if not v and items: break
        if not v: print("  ⚠ Mínimo uno."); continue
        items.append(v); i += 1
    return items

def registrar():
    print("\n" + "="*50)
    d = {}
    d["nombre"]      = pedir("Nombre del lugar")
    d["tipo"]        = pedir("Tipo de lugar (floristería, colegio, etc.)")
    d["descripcion"] = pedir("Descripción general")
    d["capacidad"]   = pedir("Capacidad o número de personas", num=True)
    d["espacios"]    = pedir("Número de espacios o ambientes", num=True)
    d["horario"]     = pedir("Horario (ej: Lun-Vie 8am-6pm)")
    d["servicios"]   = pedir_lista("Servicios o características")
    return d

def mostrar(d):
    print("\n" + "="*50)
    print(f"  {d['nombre'].upper()}  ({d['tipo']})")
    print("="*50)
    print(f"\n  {d['descripcion']}")
    print(f"\n  Capacidad : {d['capacidad']}")
    print(f"  Espacios  : {d['espacios']}")
    print(f"  Horario   : {d['horario']}")
    print(f"\n  Servicios:")
    for i, s in enumerate(d["servicios"], 1):
        print(f"    {i}. {s}")
    print("="*50)

# ── Ejecutar ───────────────────────────────────────────────
lugar = registrar()
mostrar(lugar)
`````

**POO** (Programación Orientada a Objetos) --> Creo la base pero se decora diferente

`````python
class disfraz():
  #Creando el constructor
  def __init__(self,color,talla,forma,material,personaje):
    #Creando atributos de clase
    self.color=color
    self.talla=talla
    self.forma=forma
    self.material=material
    self.personaje=personaje

  def imprimir(self):
    return f"El disfraz es de {self.personaje}.\nEs de talla {self.talla}.\nEs de material {self.material}.\nEs de color {self.color}.\nEs de forma {self.forma}."
`````

`````python
perro=disfraz("amarillo con marron","10","cuadrado", "tela", "bob esponja")
`````

`````python
perro.imprimir()
`````

`````python
print(perro.imprimir())
`````

`````python
class Edificio():
  def __init__(self,material,direccion,fachada,ventanas,puertas,nombre):
    self.material=material
    self.direccion=direccion
    self.fachada=fachada
    self.ventanas=ventanas
    self.puertas=puertas
    self.nombre=nombre

  def ubicacion(tipo):
    return f"Tipo de edificación:{tipo}\nDirección:{self.direccion}"

  def descripcion(self,tipo):
    print(f"{self.nombre} es un {tipo} ubicado en {self.direccion}")
    print(f"Tiene {self.ventanas} ventanas y {self.puertas} puerta(s)")
    print(f"Su fachada es {self.fachada}")
`````
