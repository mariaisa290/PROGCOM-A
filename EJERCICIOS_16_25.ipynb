# 18	Desarrolla un programa que calcule el factorial de un número. Recuerda que el factorial de 0 es 1 y no existen factoriales negativos.

```python
numero = int(input("Ingrese un numero: "))

factorial = 1

for i in range(1, numero + 1):
    factorial = factorial * i

if numero < 0:
    print("No existen factoriales para números negativos")
elif numero == 0:
    print("El factorial de 0 es 1")
else:
    print("El factorial de", numero, "es:", factorial)
```

```python
numero = int(input("Ingrese un numero: "))

factorial = 1

for i in range(1, numero + 1):
    if numero < 0:
        print("No existen factoriales para números negativos")
        break
    factorial = factorial * i

if numero == 0:
    print("El factorial de 0 es 1")
elif numero > 0:
    print("El factorial de", numero, "es:", factorial)
```

```python
numero = int(input("Ingrese un numero: "))

factorial = 1

for i in range(1, numero + 1, 1):   # VI=1, VF=numero, P=1
    if numero < 0:
        print("No existen factoriales para números negativos")
        break
    factorial = factorial * i

if numero == 0:
    print("El factorial de 0 es 1")
elif numero > 0:
    print("El factorial de", numero, "es:", factorial)
```

# 19 Crea un programa que muestre los primeros n números de la serie Fibonacci.

```python
n = int(input("Ingrese la cantidad de números Fibonacci que desea ver: "))

a = 0
b = 1

for i in range(n):
    if n <= 0:
        print("Debe ingresar un número positivo")
        break
    else:
        print(a)
        c = a + b
        a = b
        b = c
```

```python
n = int(input("Ingrese la cantidad de números Fibonacci que desea ver: "))

a = 0
b = 1

for i in range(n):  # cantidad de veces
    print(a)
    c = a + b
    a = b
    b = c
```

```python
n = int(input("Ingrese la cantidad de números Fibonacci que desea ver: "))

a = 0
b = 1

for i in range(1, n+1, 1):  # VI=1, VF=n, Paso=1
    print(a)
    c = a + b
    a = b
    b = c
```

# 23 Escribe un programa que determine si un número es primo.

```python
numero = int(input("Ingrese un número: "))

for i in range(2, numero):
    if numero <= 1:
        print("No es un número primo")
        break
    elif numero % i == 0:
        print("No es un número primo")
        break
else:
    if numero > 1:
        print("Es un número primo")
```

```python
numero = int(input("Ingrese un número: "))

for i in range(numero):
    if numero <= 1:
        print("No es un número primo")
        break
    elif i >= 2 and numero % i == 0:
        print("No es un número primo")
        break
else:
    print("Es un número primo")
```

```python
numero = int(input("Ingrese un número: "))

for i in range(1, numero + 1, 1):  # VI=1, VF=numero, P=1
    if numero <= 1:
        print("No es un número primo")
        break
    elif i > 1 and i < numero and numero % i == 0:
        print("No es un número primo")
        break
else:
    print("Es un número primo")
```
